# Petite demo audio

J'utilise `numpy` pour travailler facilement avec des tableaux et `matplotlib` pour afficher les signaux.

Pour lire les signaux audio, j'utilise simplement l'objet `Àudio`du module `IPython` (module au coeur de jupyter)

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
# médium
f0 = 1000
Fe = 44100
t = np.arange(0,2,1/Fe)
x = np.sin(2*np.pi*f0*t)

plt.plot(t,x)
plt.xlim([0,0.01])
plt.xlabel("temps (s)");
```

```python
Audio(x,rate=Fe)
```

```python
# La
f0 = 220
Fe = 44100
t = np.arange(0,2,1/Fe)
x = np.sin(2*np.pi*f0*t)

plt.plot(t,x)
plt.xlim([0,0.01])
plt.xlabel("temps (s)");
```

```python
list_a = [1,0.2,0.1,0.2,0.125,0.1]
list_phi= [0,0.1,0.2,-0.3,0.2,0.1]
f0 = 220

t = np.arange(0,2,1/Fe)
x = np.zeros(len(t))

# je vais rajouter les harmonique au fur et à mesure
for n in range(len(list_a)):
    f = (n+1)*f0   # attention fondamentale en en f0 pas en 0*f0
    x += list_a[n]*np.cos(2*np.pi*f*t+list_phi[n])
    
plt.plot(t,x)
plt.xlim([0,0.01])
plt.xlabel("temps (s)");
Audio(x,rate=Fe)
```
